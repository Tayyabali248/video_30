```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
kashti=sns.load_dataset("titanic")
sns.boxplot(x="survived", y="age", data=kashti)
plt.show()
```

```python
sns.boxplot(x="survived", y="age", showmeans=True, data=kashti)
plt.show()
```

```python
sns.boxplot(x="survived",
            y="age", showmeans=True,
            meanprops={"marker": "*",
                      "markersize": "14",
                      "markeredgecolor": "red"}, data=kashti)
plt.xlabel("zinda",size=12)
plt.ylabel("umer(year)",size=14)
plt.title("weshi",size=20,weight='bold')
plt.show()
```
